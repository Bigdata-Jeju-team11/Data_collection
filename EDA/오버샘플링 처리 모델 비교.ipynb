불균형한 상태에서 모델링 진행 및 성능 평가하기
- 각 컬럼에서 데이터의 개수가 현저하게 차이가 난 상태로 모델 학습시, 다수의 범주로 패턴 분류를 많이 하게 되는 문제가 발생
-> 하지만, 현재 DB는 결측값을 제거한 데이터이며, (범주형 변수)해당 하천 수위가 교량 높이를 넘으면 침수된 것이므로 과적합이 일어난다.
- 이는 곧 모델의 성능에 영향을 끼친다.
-> 범주형 변수를 사용하는 모델은 성능에 영향을 주지 않고, 과적합 발생

```python
## 여기에 모델링 해보기
```

```python
from sklearn.model_selection import train_test_split

X = df[df['station_code']==1007604].iloc[:-1, [6, 9, -1]].values # iloc[row, column] => rainfall: 6, wl: 9
Y = df[df['station_code']==1007604]['Flooding'].iloc[:-1].values
# len(X), len(Y)
#X=X.reshape(-1,1)
#Y=Y.reshape(-1,1)
X_train, X_test, Y_train, Y_test = train_test_split(X,Y,test_size=0.25,random_state=10)
Y_test, Y_train, X, Y
```

```python
def modeling(model,x_train,x_test,y_train,y_test):
    model.fit(x_train,y_train.ravel())
    pred = model.predict(x_test)
    metrics(y_test,pred)
#평가 지표
def metrics(y_test,pred):
    accuracy = accuracy_score(y_test,pred)
    precision = precision_score(y_test,pred, average='micro')
    recall = recall_score(y_test,pred, average='micro')
    f1 = f1_score(y_test,pred, average='micro')
    roc_score = roc_auc_score(y_test, pred, average='micro')#, multi_class='ovo')
    print('정확도 : {0:.2f}, 정밀도 : {1:.2f}, 재현율 : {2:.2f}'.format(accuracy,precision,recall))
    print('f1-score : {0:.2f}, auc : {1:.2f}'.format(f1,roc_score,recall))
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
from sklearn.metrics import f1_score
from sklearn.metrics import roc_auc_score

lr = LogisticRegression(solver='lbfgs')#, max_iter=10000)
modeling(lr,X_train,X_test,Y_train,Y_test)
```

## RandomOverSampler, SMOTE 적용
- 수집한 데이터는 생각보다 이상적이지 않다.
- 불균형 데이터 문제를 해결할 수 있는 smote 적용: 동작방식은 데이터의 개수가 적은 컬럼의 표본을 가져온 뒤 임의의 값을 추가하여 새로운 샘플을 만들어 데이터에 추가하는 오버샘플링 방식이다.
- imbalanced 한 데이터 처리 후, 모델링 및 성능 평가하기

```python
!pip install imblearn
```

```python
from imblearn.over_sampling import SMOTE
from imblearn.over_sampling import RandomOverSampler
```

```python
# oversample = RandomOverSampler(sampling_strategy='minority')
# X_train_over, Y_train_over = oversample.fit_resample(X_train, Y_train)
smote = SMOTE(random_state=0, sampling_strategy='minority')
X_train_over, Y_train_over = smote.fit_resample(X_train, Y_train)
```

```python
print('OverSampler 적용 전 학습용 피처/레이블 데이터 세트: ', X_train.shape, Y_train.shape)
print('OverSampler 적용 후 학습용 피처/레이블 데이터 세트: ', X_train_over.shape, Y_train_over.shape)
print('OverSampler 적용 후 레이블 값 분포: \n', pd.Series(Y_train_over).value_counts())
```

```python
lr = LogisticRegression(solver='lbfgs')#, max_iter=10000)
modeling(lr,X_train_over,X_test,Y_train_over,Y_test)
```
