# Isolation Forest

```python
df.corr()
```

```python
station_code_list = pd.unique(df['station_code']).tolist()
```

```python
test_df = df.drop(columns=['Unnamed: 0', 'obs_date_wl', 'station_code', 'station_name', 'obs_date_rf', 'pp_station', 'rainfall',
                           'bridge_name', 'brid_height_origin', 'bridge_height'])
```

```python
test_df
```

```python
df_list = [df[df['station_code']==i] for i in station_code_list]
```

```python
test_list = [df_list[i].drop(columns=['Unnamed: 0', 'obs_date_wl', 'station_code', 'station_name', 'obs_date_rf', 'pp_station',
                               'rainfall', 'bridge_name', 'brid_height_origin', 'bridge_height']) for i in range(len(df_list))]
```

```python
from sklearn.ensemble import IsolationForest
```

```python
clf=IsolationForest(n_estimators=50, max_samples=50, contamination=float(0.004), 
                    max_features=1.0, bootstrap=False, n_jobs=-1, random_state=None, verbose=0,behaviour="new")
```

```python
for i in range(len(test_list)):
    clf.fit(test_list[i])
```

```python
pred_list = [clf.predict(test_list[i]) for i in range(len(test_list))]
```

```python
for i in range(len(test_list)):
    test_list[i]['anomaly'] = pred_list[i]
```

```python
outliers_list = [test_list[i].loc[test_list[i]['anomaly']==-1] for i in range(len(test_list))]
```

```python
outlier_idx_list = [outliers_list[i].index for i in range(len(outliers_list))]
```

```python
outlier_idx_list[0]
```

```python
print(outlier_idx_list)
```

```python
for i in range(len(test_list)):
    print('df', i)
    print(test_list[i]['anomaly'].value_counts())
```

```python
import numpy as np
from sklearn.decomposition import PCA
```

```python
pca = PCA(2)
```

```python
pca_list = [pca.fit(test_list[i]) for i in range(len(test_list))]
```

```python
res_list = [pd.DataFrame(pca.transform(test_list[i])) for i in range(len(test_list))]
```

```python
Z_list = [np.array(res_list[i]) for i in range(len(res_list))]
```

```python
import matplotlib.pyplot as plt
```

```python
plt.rcParams["figure.figsize"] = (14,4)
```

```python
start_idx = 0
for i in range(len(res_list)):
    plt.title('Isolation Forest')
    b1 = plt.scatter(res_list[i][0], res_list[i][1], c='green', s=20, label='normal points')
    b1 = plt.scatter(res_list[i].iloc[(outlier_idx_list[i]-start_idx),0], res_list[i].iloc[(outlier_idx_list[i]-start_idx),1],
                    c='green', s=20, edgecolor='red', label='predicted outliers')
    plt.legend(loc='upper right')
    plt.show()
    start_idx += len(test_list[i])
```
