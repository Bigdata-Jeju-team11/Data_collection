```python
import geopandas as gdp
```

```python
import matplotlib.pyplot as plt
```

```python
plt.rcParams["font.family"] = 'NanumGothic'
plt.rcParams["figure.figsize"] = (10,10)
```

```python
fl_file = 'D:/flood_data/fl_shelter/FL_RESCUE_SP.shp'
```

```python
fl = gdp.read_file(fl_file)
```

```python
fl
```

```python
fl1 = fl[fl['SIDO_NM']=='서울특별시']
```

```python
ax = fl1.convex_hull.plot(color='blue', edgecolor='w')
ax.set_title("서울특별시 침수 대피소")
ax.set_axis_off() # 축 제거
plt.show()
```
