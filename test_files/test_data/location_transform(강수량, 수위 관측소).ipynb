# 강수량, 수위 관측소
## 좌표(위도, 경도) 변환 코드
- 관측소 데이터에 DD 방식 컬럼 추가
- GPS의 정보의 경위도 좌표계의 표시방법 2가지:DD(Decimal Degree) 와 DMS(Degree Minutes Seconds) 방식
- 변환식은 DD = Degree + Minutes/60 + Seconds/3600
- Latitude(위도) => 36°27′08.0″N  -> 36.45222
- Longitude(경도) => 127°30′20.0″E -> 127.50556

```python
import pandas as pd
```

```python
df1 = pd.read_csv('C:\\jupyter_notebook_file\\NoonMap_origin\\NoonMap-Project\\data\\file_data\\수위_관측소.csv')
df2 = pd.read_csv('C:\\jupyter_notebook_file\\NoonMap_origin\\NoonMap-Project\\data\\file_data\\강수량_관측소_api.csv')

df1['tr_lat'] = 0
df1['tr_long'] = 0
df2['tr_lat'] = 0
df2['tr_long'] = 0
```

```python
#위도경도 변환 함수
#위경도 변환 DD = Degree(앞 3자리) + Minutes(중간 2자리)/60 + Seconds/3600
def transform_location(frame, i):
        tmp_lat = frame['latitude'].loc[i].split('-')
        transform_lat = int(tmp_lat[0])+int(tmp_lat[1])/60+int(tmp_lat[2])/3600

        tmp_long = frame['longitude'].loc[i].split('-')
        transform_long = int(tmp_long[0])+int(tmp_long[1])/60+int(tmp_long[2])/3600
        
        frame['tr_lat'].loc[i] = transform_lat
        frame['tr_long'].loc[i] = transform_long
        
for i in range(len(df1)):
    transform_location(df1, i)
    
for i in range(len(df2)):
    transform_location(df2, i)
```
