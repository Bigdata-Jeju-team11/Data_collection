```python
station_code_list = ["1015310", "1017310", "1021701", "1022701"]
```

```python
def dam_api():
    url = 'http://api.hrfco.go.kr/A5F02E26-FBD3-47F7-A191-280B5944845D/dam/list/1H.json'
    
    req = requests.get(url)
    
    json_row = req.json()['content']
    
    url_list = ['http://api.hrfco.go.kr/A5F02E26-FBD3-47F7-A191-280B5944845D/dam/list/1H/' + i for i in station_code_list]
    
    date = [ '/20' + str(i) + '010100/20' + str(i) + '123123.json' for i in range(10,21) ]
    
    code = []
    YMDHM = []
    SWL = []
    INF = []
    SFW = []
    ECPC = []
    TOTOTF = []
    
    for i in tqdm(url_list):
        for j in date:
            url = i + j
            req = requests.get(url)
            json_row = req.json()['content']
            
            for k in range(len(json_row)):
                row = json_row[k]
                code.append(row['dmobscd'])
                YMDHM.append(row['ymdhm'])
                SWL.append(row['swl'])
                INF.append(row['inf'])
                SFW.append(row['sfw'])
                ECPC.append(row['ecpc'])
                TOTOTF.append(row['tototf'])
                
    df = pd.DataFrame({'station_code':code, 'obs_date':YMDHM, '현재 수위':SWL, '유입량':INF,
                      '저수량':SFW, '공용량':ECPC, '총_방류량':TOTOTF})
    
    return df
```

```python
dam_df = dam_api()
```
