```python
station_code_list = pd.unique(df['station_code']).tolist()
```

```python
df_list = []
X = []
y = []
for i in range(len(station_code_list)):
    df_list.append(df[df['station_code']==station_code_list[i]])
    model = LinearRegression()
    X.append(df_list[i][['rainfall']])
    y.append(df_list[i]['WL'])
    model.fit(X[i], y[i])
    print( "model:", i)
    print( "coefficient=", model.coef_ )
    print( "intercept=", model.intercept_ )
    print( "R²=", model.score(X[i], y[i]) )
    plt.scatter(X[i], y[i])
    plt.show()
    print()
```
