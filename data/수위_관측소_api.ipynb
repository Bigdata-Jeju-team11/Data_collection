```python
def suwi_station():
    url = 'http://api.hrfco.go.kr/A5F02E26-FBD3-47F7-A191-280B5944845D/waterlevel/info.json'
    
    req = requests.get(url)
    
    json_row = req.json()['content']
    
    code = []
    agency = []
    name = []
    addr = []
    etcaddr = []
    lon = []
    lat = []
    gdt = []
    wrnwl = []
    almwl = []
    pfh = []
    
    for i in range(len(json_row)):
        row = json_row[i]
        code.append(row['wlobscd'])
        agency.append(row['agcnm'])
        name.append(row['obsnm'])
        addr.append(row['addr'])
        etcaddr.append(row['etcaddr'])
        lon.append(row['lon'])
        lat.append(row['lat'])
        gdt.append(row['gdt'])
        wrnwl.append(row['wrnwl'])
        almwl.append(row['almwl'])
        pfh.append(row['pfh'])
        
    df = pd.DataFrame({"station_code": code, "agency_name": agency, "station_name":name, "address":addr, "etc_address":etcaddr,
                       "longitude": lon, "latitude": lat, "zero_elevation": gdt, "warning_wl": wrnwl, "alarm_wl": almwl})
    
    return df
```

```python
suwi_station_df = suwi_station()
```
