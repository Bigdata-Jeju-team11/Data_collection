```python
#위도경도 변환 함수

#위경도 변환 DD = Degree(앞 3자리) + Minutes(중간 2자리)/60 + Seconds/3600

def location(x,i: int,list_name):
    list0 = str(x['longitude'][i]).split('-')
    list1 = str(x['latitude'][i]).split('-')
    DD1 = int(list0[0]) + int(list0[1])/60 + int(list0[2])/3600
    DD2 = int(list1[0]) + int(list1[1])/60 + int(list1[2])/3600
    list_name.append([DD1,DD2])
```

```python
for i in range(0, len(df_wl_name)):
    location(df_wl_name,i, loc_list_wl)
```

# 수위 관측소와 가장 가까운 강수량 관측소 찾기
수위 관측소의 각 위치(df_wl_name['location'])마다 df_pp_name['location'] 전체와의 거리 차이를 구한다. 거리 차이가 가장 작은 관측소 코드를 'pp_station_code'라는 변수로 설정하여 수위 데이터에 추가한다.

```python
pp_station = []

def closest_station(x):
    for i in range(0,len(df_wl_name)):
        min_index = 0
        for j in range(0,len(df_pp_name)):
            a = haversine(df_wl_name['location'][i],df_pp_name['location'][j])
            if j == 0:
                min_dist = a
            else:
                if a<=min_dist:
                    min_dist= a
                    min_index = j
        pp_station.append(df_pp_name['station_code'][min_index])

                
```
