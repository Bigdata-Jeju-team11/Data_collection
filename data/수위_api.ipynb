```python
def suwi_api():
    url = 'http://api.hrfco.go.kr/A5F02E26-FBD3-47F7-A191-280B5944845D/waterlevel/list/1H.json'
    
    req = requests.get(url)
    
    json_row = req.json()['content']

    station_code = [ json_row[i]['wlobscd'] for i in range(len(json_row)) ]
    
    url_list = [ 'http://api.hrfco.go.kr/A5F02E26-FBD3-47F7-A191-280B5944845D/waterlevel/list/1H/' + i for i in station_code ]
    
    date = [ '/20' + str(i) + '010100/20' + str(i) + '123123.json' for i in range(10,21) ]
    
    code = []
    YMDHM = []
    WL = []
    FW = []
    
    for i in tqdm(url_list):
        for j in date:
            url = i + j
            req = requests.get(url)
            json_row = req.json()['content']
            
            for k in range(len(json_row)):
                row = json_row[k]
                code.append(row['wlobscd'])
                YMDHM.append(row['ymdhm'])
                WL.append(row['wl'])
                FW.append(row['fw'])
                
    df = pd.DataFrame({'station_code':code, 'obs_date':YMDHM, 'WL':WL, 'FW':FW})
    
    return df
```

```python
suwi_df = suwi_api()
```
