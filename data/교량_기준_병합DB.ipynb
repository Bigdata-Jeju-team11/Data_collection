## 수위 관측소와 가장 가까운 교량 매칭

```python
location_start = [[gyo["latitude"][i], gyo["longitude"][i]] for i in range(len(gyo))]
```

```python
gyo["location_start"] = location_start
```

```python
wl_station = pd.read_csv(r'C:\Users\Shin\Desktop\for-kr-pjt\NoonMap-Project\data\file_data\수위_관측소.csv')
```

```python
wl_station.drop(['Unnamed: 0'], axis=1, inplace=True)
```

```python
wl_station
```

```python
def location(x,i: int,list_name):
    list0 = str(x['latitude'][i]).split('-')
    list1 = str(x['longitude'][i]).split('-')
    DD1 = int(list0[0]) + int(list0[1])/60 + int(list0[2])/3600
    DD2 = int(list1[0]) + int(list1[1])/60 + int(list1[2])/3600
    list_name.append([DD1,DD2])
```

```python
loc_list_wl = []
for i in range(0, len(wl_station)):
    location(wl_station, i, loc_list_wl)
```

```python
wl_station["location"] = loc_list_wl
```

```python
from haversine import haversine
```

```python
brid_location = []
def closest_station(x):
    for i in range(len(wl_station)):
        min_index = 0
        for j in range(len(gyo)):
            a = haversine(wl_station['location'][i], gyo['location_start'][j])
            if j == 0:
                min_dist = a
            else:
                if a<=min_dist:
                    min_dist= a
                    min_index = j
        brid_location.append(gyo['location_start'][min_index])
```

```python
closest_station(wl_station)
```

# 교량 데이터의 교량 높이 컬럼 보완
- 교량 높이 측정 기준: 교량 바닥판 최상면에서 지표면 또는 수면까지의 수직 방향 최장거리
- 수면의 높이를 구하기 위해 하천별로 수위 관측값의 최빈값을 사용

```python
df = pd.read_csv('병합DB_1차.csv')
```

```python
brid_list = pd.unique(df['location_start']).tolist() # 교량 위치별로
```

```python
len(brid_list)
```

```python
df_list = [df[df['location_start']==i] for i in brid_list]
```

```python
for i in range(len(df_list)):
    print(df_list[i]['WL'].mode())
```

```python
float(df_list[0]['WL'].mode()[0])
```

```python
wl_mode = [float(df_list[i]['WL'].mode()[0]) for i in range(len(df_list))]
```

```python
len(wl_mode)
```

```python
mode_list = []
for i in range(len(df_list)):
    for j in range(len(df_list[i])):
        mode_list.append(wl_mode[i])
```

```python
df['bridge_height'] = (df['brid_height_origin'] * 100) + mode_list # cm로 단위 통일
```
