```python
def rain_station():
    url = 'http://api.hrfco.go.kr/A5F02E26-FBD3-47F7-A191-280B5944845D/rainfall/info.json'
    
    req = requests.get(url)
    
    json_row = req.json()['content']
    
    code = []
    agency = []
    name = []
    addr = []
    etcaddr = []
    lon = []
    lat = []

    for i in range(len(json_row)):
        row = json_row[i]
        code.append(row['rfobscd'])
        agency.append(row['agcnm']) 
        name.append(row['obsnm'])
        addr.append(row['addr'])
        etcaddr.append(row['etcaddr'])
        lon.append(row['lon'])
        lat.append(row['lat'])
        
    df = pd.DataFrame({"station_code": code, "agency_name": agency, "station_name":name, "address":addr, "etc_address":etcaddr,
                       "longitude": lon, "latitude": lat})
    
    return df
```

```python
rain_station_df = rain_station()
```
