```python
df_station = pd.read_csv(r'C:\Users\Shin\Desktop\NoonMap-Project\data\file_data\수위_관측소.csv')
```

```python
merge_df = pd.merge(df_station, df, on='station_code')
```

```python
merge_df.drop(columns = ["Unnamed: 0_x", "longitude", "latitude", "Unnamed: 0_y"], axis=1, inplace=True)
```

```python
station_code_list = pd.unique(merge_df['station_code']).tolist()
```

```python
df_list = []
for i in range(120):
    i = pd.DataFrame()
    df_list.append(i)
```

```python
for i in range(120):
    df_list[i] = merge_df[merge_df['station_code'] == station_code_list[i]]
```

```python
df_list[6]
```

```python
zero_list = [7, 8, 10, 12, 27, 70, 82, 83, 85, 90, 91]
```

```python
idx_list = []
for i in zero_list:
    idx_list.append(df_list[i][df_list[i]["WL"] == 0.0].index)
```

```python
for i in range(9):
    df_list[zero_list[i]] = df_list[zero_list[i]].drop(idx_list[i])
```

```python
del_list = [6, 14, 16, 21, 26, 29, 30, 37, 41, 67, 68, 69, 78, 84, 101, 103, 105, 113, 114, 115, 116, 117, 118, 119]
```

```python
frame = []
for i in range(120):
    if i not in del_list:
        frame.append(df_list[i])
```

```python
frame
```

```python
final_df = pd.concat(frame)
```

```python
final_df
```

```python
final_df.to_csv("수위_최종.csv")
```

```python
test = pd.read_csv("수위_최종.csv")
```

```python
station_code_list = pd.unique(test['station_code']).tolist()
```

```python
len(station_code_list)
```

```python
tt_list = []
for i in range(96):
    i = pd.DataFrame()
    tt_list.append(i)
```

```python
for i in range(96):
    tt_list[i] = test[test['station_code'] == station_code_list[i]]
```

```python
import matplotlib.pyplot as plt
```

```python
plt.rc('font', family='Malgun Gothic')
```

```python
station_name_list = pd.unique(test['station_name']).tolist()
```

```python
for i in range(96): # 각 관측소마다
    plt.xlabel('obs_datetime') # 시간의 흐름에 따른
    plt.ylabel('WL') # 하천 수위의 변화
    plt.title(str(i) + '  ' + station_name_list[i])
    plt.plot(tt_list[i]['WL'])
    plt.show()
```

```python
# 전후 흐름과 맞지 않게 갑자기 0의 관측값을 가지는 데이터들이 있음
# 측정 기기 오류로 인한 결측값이라 판단하고 제거

zero_list1 = [0, 22, 59, 60, 61, 62]
```

```python
idx_list1 = []
for i in zero_list1:
    idx_list1.append(tt_list[i][tt_list[i]["WL"] == 0.0].index)
```

```python
for i in range(6):
    tt_list[zero_list1[i]] = tt_list[zero_list1[i]].drop(idx_list1[i])
```

```python
frame1 = []
for i in range(96):
    frame1.append(tt_list[i])
```

```python
final_df1 = pd.concat(frame1)
```

```python
final_df1.to_csv("수위_진짜_최종.csv")
```
