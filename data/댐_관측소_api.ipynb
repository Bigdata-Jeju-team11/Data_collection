```python
def dam_station():
    url = 'http://api.hrfco.go.kr/A5F02E26-FBD3-47F7-A191-280B5944845D/dam/info.json'
    
    req = requests.get(url)
    
    json_row = req.json()['content']
    
    code = []
    name = []
    agency = []
    addr = []
    etcaddr = []
    lon = []
    lat = []
    pfh = []
    fldlmtwl = []
    
    for i in range(len(json_row)):
        row = json_row[i]
        
        if row == None:
            pass
        else:
            code.append(row['dmobscd'])
            name.append(row['obsnm'])
            agency.append(row['agcnm']) 
            addr.append(row['addr'])
            etcaddr.append(row['etcaddr'])
            lon.append(row['lon'])
            lat.append(row['lat'])
            pfh.append(row['pfh'])
            fldlmtwl.append(row['fldlmtwl'])
        
    df = pd.DataFrame({"station_code": code, "agency_name": agency, "station_name":name, "address":addr, "etc_address":etcaddr,
                       "longitude": lon, "latitude": lat, "pfh": pfh, "fldlmtwl": fldlmtwl})
    
    return df
```

```python
dam_station_df = dam_station()
```
