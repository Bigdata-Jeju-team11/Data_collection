```python
df_station = pd.read_csv(r'C:\Users\Shin\Desktop\for-kr-pjt\NoonMap-Project\data\file_data\수위_관측소.csv')
```

```python
df_station.drop(['Unnamed: 0'], axis=1, inplace=True)
```

```python
# 위경도 좌표 변환 컬럼 추가
def location(x,i: int,list_name):
    list0 = str(x['latitude'][i]).split('-')
    list1 = str(x['longitude'][i]).split('-')
    DD1 = int(list0[0]) + int(list0[1])/60 + int(list0[2])/3600
    DD2 = int(list1[0]) + int(list1[1])/60 + int(list1[2])/3600
    list_name.append([DD1,DD2])
```

```python
loc_list_wl = []
for i in range(len(df_station)):
    location(df_station, i, loc_list_wl)
```

```python
df_station["location"] = loc_list_wl
```

```python
merge_df = pd.merge(df_station, df, on='station_code')
```

```python
# 관측일자 컬럼의 형식과 데이터 타입 변환
merge_df['obs_date'] = pd.to_datetime(merge_df['obs_date'], format = '%Y%m%d%H')
```

```python
# 불필요한 컬럼 제거
merge_df.drop(columns=['longitude', 'latitude'], inplace=True)
```
