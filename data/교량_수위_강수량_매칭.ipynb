```python
# 근데 이게 맞나...? 아직 모델에 대한 이해도가 부족함
# Y: 침수 여부 == (특정 교량의 높이 - 해당하는 하천의 예상 수위) 1. < 0: 침수O / 2. >= 0: 침수X
# Y: 침수 여부 -> 이 모델을 위해서 Y: 하천의 예상 수위 -> 이 모델을 활용해야 함
# Y: 하천의 예상 수위 = 해당 하천과 가장 가까운 강수량 관측소의 강수량 + 해당 하천에 영향을 주는 댐의 방류량
# 근데 과거 데이터면 몰라도 실시간 강수량을 관측소마다 미리 알 수 있나?

# 특정 교량의 1시간 뒤 침수 여부 == (특정 교량의 높이 - 특정 교량에 해당하는 1시간 뒤 예상 하천 수위)
# 특정 교량에 해당하는 강수량 + 특정 교량에 영향을 미치는 댐의 방류량

# 특정 하천의 특정 시간의 수위 = 특정 하천과 가까운 관측소의 특정 시간의 강수량 + 특정 하천에 영향을 미치는 댐의 특정 시간의 방류량
# 다중선형회귀분석을 통해 강수량과 특정 댐의 방류량이 하천 수위에 어떻게, 얼마큼 영향을 미치는지 확인(아마 양의 선형 관계)
# 근데 잠시만... 1시간 뒤 예상 하천 수위는 그러면 어떻게 구하나?
# 1시간 뒤 예상 강수량이면 몰라도 예상 댐의 방류량은 우리가 알 수 없음
# 그렇다면 하천 수위 테이블과 강수량 테이블/댐 테이블은 병합할 때 1시간의 시간차를 두어야 함
# 특정 하천의 1시간 뒤 예상 수위 = 특정 하천과 가까운 관측소의 현재 강수량 + 특정 하천에 영향을 미치는 댐의 현재 방류량
```
