```python
def rainfall_api():
    url = 'http://api.hrfco.go.kr/A5F02E26-FBD3-47F7-A191-280B5944845D/rainfall/list/1H.json'
    
    req = requests.get(url)
    
    json_row = req.json()['content']
    
    url_list = ['http://api.hrfco.go.kr/A5F02E26-FBD3-47F7-A191-280B5944845D/rainfall/list/1H/' + i for i in station_code_list]
    
    date = [ '/20' + str(i) + '010100/20' + str(i) + '123123.json' for i in range(10,21) ]
    
    code = []
    YMDHM = []
    rf = []
    
    for i in tqdm(url_list):
        for j in date:
            url = i + j
            req = requests.get(url)
            json_row = req.json()['content']
            
            for k in range(len(json_row)):
                row = json_row[k]
                code.append(row['rfobscd'])
                YMDHM.append(row['ymdhm'])
                rf.append(row['rf'])
                
    df = pd.DataFrame({'station_code':code, 'obs_date':YMDHM, 'rainfall':rf})
    
    return df
```

```python
rainfall_df = rainfall_api()
```
