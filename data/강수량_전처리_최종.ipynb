#### 강수량 관측소와 강수량 데이터 병합

```python
merge_df = pd.merge(df_station, df, on='station_code')
```

```python
merge_df
```

```python
merge_df.info()
```

```python
# 관측소 갯수, 관측소별 데이터 갯수
total = 0
for i in range(len(sudo_station_list)):
    total += len(merge_df[merge_df['station_code']==sudo_station_list[i]])
    print(sudo_station_list[i], len(merge_df[merge_df['station_code']==sudo_station_list[i]]))
    merge_df[merge_df['station_code']==sudo_station_list[i]].sort_values(by='obs_date')
    
print(len(sudo_station_list), len(merge_df), total)
```

```python
## 수도권 지역 리스트
address_list = pd.unique(df_station['address']).tolist()
address_list, len(address_list)
```

```python
## 지역별 관측소 리스트
stations_by_address_list = []
total = 0
for i in range(len(address_list)):
    total += len(merge_df[merge_df['address']==address_list[i]])
    stations = pd.unique(merge_df[merge_df['address']==address_list[i]]['station_code']).tolist()
    stations_by_address_list.append(stations)
    print(pd.unique(merge_df[merge_df['address']==address_list[i]]['station_code']).tolist())
    print(address_list[i], len(merge_df[merge_df['address']==address_list[i]]))
    
print(total)
stations_by_address_list
```

```python
len(stations_by_address_list)
```

# 지역별 관측소마다의 강수량

```python
import matplotlib.pyplot as plt
import matplotlib as mpl
```

```python
plt.rc('font', family='Malgun Gothic')
mpl.rcParams['axes.unicode_minus'] = False
```

```python
for i in range(len(stations_by_address_list)):
    #plt.figure(figsize=(18,10))
    for j in range(len(stations_by_address_list[i])):
        plt.figure(figsize=(18,10))
        plt.xlabel('년도')
        plt.ylabel('강수량')
        plt.title(address_list[i] + ', 관측소 갯수: ' + str(len(stations_by_address_list[i])) + ', 관측소 코드: ' + str(stations_by_address_list[i][j]))
        plt.plot(merge_df[merge_df['station_code']==stations_by_address_list[i][j]]['obs_date'], merge_df[merge_df['station_code']==stations_by_address_list[i][j]]['rainfall'], label=str(stations_by_address_list[i][j]))
        plt.show()
```
